```python
video_name = Path("Video_4_Poor")
videos_path = new_videos_path / video_name
videos = glob.glob( str(videos_path) + "/*.mp4")

# min * fps * conv.
init1 = 2/3 * 30 * 60
end1 = (1+1/6) * 30 * 60

init2 = 0 * 30 * 60
end2 = 0 * 30 * 60

save_path = "Visibility_levels_test/5_Very_Poor/"
check_dir(Path(save_path))


for i,video in enumerate(videos):
    vidcap = cv2.VideoCapture(video)
    success,image = vidcap.read()
    count = 0
    while success:
        if end1 > count > init1 or end2 > count > init2:
            frame_path_mod = save_path + str(video_name) + "_" +str(i) +"_frame%d.png" % count
            cv2.imwrite(frame_path_mod, image)     # save frame as JPEG file

        success,image = vidcap.read()

        if(count % 500 == 0):
            print(f"Frame:{count} | Success:{success}")
        count += 1
```
