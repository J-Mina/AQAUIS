```python
#Transform Quality issue videos into frames (without any selection)
root = Path('Quality/raw/')
list_path = list(root.iterdir())
dest_dir = Path('Quality/dataset/')
for path in list_path:
    problems = root / Path(path.stem + "/")
    videos_path = list(problems.iterdir())
    for video_path in videos_path:
        videos = glob.glob(str(video_path) + "\*.mp4")
        for i, video in enumerate(videos):
            video_name = video_path.parent.stem + "/" + video_path.stem + "_" + str(i)
            video_classes = dest_dir / Path(video_path.parent.stem + "/")
            check_dir(video_classes)
            video_name = str(dest_dir / video_name)
            convert_mp4_to_png(video, video_name)
```
