```python
folder_path = Path("spaced_data_separated")
vp_cnt = 0
p_cnt = 0
m_cnt = 0
g_cnt = 0

for i,video in enumerate(list(folder_path.iterdir())):
    
    for quality in list(video.iterdir()):
        files = list(quality.glob('*'))
        file_count = len(files)

        name = quality.stem

        if name == "Very Poor":
            vp_cnt += file_count
        elif name == "Poor":
            p_cnt += file_count
        elif name == "Moderate":
            m_cnt += file_count
        elif name == "Good":
            g_cnt += file_count

print("Good:", g_cnt ," / Moderate:", m_cnt, " / Poor:", p_cnt, " / Very Poor:", vp_cnt)
```
